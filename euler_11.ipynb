```python
# globals
max_product = 0
max_sample = [0,0,0,0]
```

```python
def sample_product(s):
    ''' returns the product of the elements in s '''
    global max_product
    product = s[0] * s[1] * s[2] * s[3]

    if product > max_product:
        max_product = product

    return product

```

```python
def main():
    global max_product
    for x in range(GRID_SIZE-ADJ_SIZE+1):
        for y in range(GRID_SIZE-ADJ_SIZE+1):

            print('Horizontal   {}  {}'.format(x, y))
            sample = [grid[y][x], grid[y+1][x], grid[y+2][x], grid[y+3][x]]
            print('{}    {}'.format(sample, sample_product(sample)))

            print('Vertical')
            sample = [grid[y][x], grid[y][x+1], grid[y][x+2], grid[y][x+3]] 
            print('{}    {}'.format(sample, sample_product(sample)))

            print('Diagonal')
            sample = [grid[y][x], grid[y+1][x+1], grid[y+2][x+2], grid[y+3][x+3]] 
            print('{}    {}'.format(sample, sample_product(sample)))

            print('Reverse Diagonal')
            sample = [grid[y][x+3], grid[y+1][x+2], grid[y+2][x+1], grid[y+3][x]] 
            print('{}    {}'.format(sample, sample_product(sample)))
    print('Max = {}   {}'.format(max_sample, max_product))

```

```python
def timing_function(timed_function):
    ''' Outputs time a function takes to execute '''
    def wrapper():
        ''' wrapper '''
        t_start = time.time()
        timed_function()
        t_end = time.time()
        return '\nScript ran for {} seconds\n'.format(str((t_end - t_start)))
    return wrapper
```
