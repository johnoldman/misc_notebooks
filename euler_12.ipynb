```python
def calc_if_prime(number):
    '''return true if number is prime'''
    if number == 1: return 0
    if number == 2: return 1
    if number % 2 == 0: return 0 # Even number 
    for i in range(2, number):
        # Advance from two to include correct calculation for '4'
        if number % i == 0: return 0
    return 1
```

```python
def print_divisors(num):
    '''print devisors of a number'''
    i = 1
    while i <= num : 
        if (num % i==0) : 
            print (i ,)
        i = i + 1
```

```python
def count_divisors(num):
    '''count devisors of a number'''
    count = 0
    i = 1
    while i <= num : 
        if (num % i==0) :
            count += 1
        i = i + 1
    return count
```

```python
def triangle_num_rule(num):
    ''' return triangle number '''
    return int((num*(num + 1))/2)
```

```python
def primes_in_num(num):
    ''' return number of primes in a number''' 
    primes = 1
    for i in range(1, (num + 1)):
        if calc_if_prime(i):
            print('{} Is Prime'.format(i))
            primes += 1;
    print('{} primes in {}'.format(primes, num))
    return primes
```
